```python
# 강사님 풀이
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data_iris.csv')

x = df['SepalLength']
y = df['PetalLength']

# 다양한 슬라이싱 방법들
# x = df.SepalLength
# x = df.loc[:, 'SepalLength']

df.Species.value_counts()
```

```python
df2 = df[(df.SepalLength >= 4)& (df.SepalLength <= 7)]

df_setosa = df2[df2.Species == 'setosa']
df_virginica = df2[df2.Species == 'virginica']
df_versicolor = df2[df2.Species == 'versicolor']
```

```python
plt.scatter(df_setosa.SepalLength, df_setosa.PetalLength, c='g', marker='o', alpha=0.3, label='Setosa')
plt.scatter(df_virginica.SepalLength, df_virginica.PetalLength, c='r', marker='o', alpha=0.3, label='Virginica')
plt.scatter(df_versicolor.SepalLength, df_versicolor.PetalLength, c='b', marker='o', alpha=0.3, label='Versicolor')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('iris flower')
plt.legend()
plt.show()
```
