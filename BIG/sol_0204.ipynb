#### 필요한 부분을 슬라이싱 해서 가져온다:

```python
# west=df.loc[df.Region =='West','Units']
# central = df.loc[df.Region =='Central','Units']
# east = df.loc[df.Region =='East','Units']
```

```python
west = df[df.Region == 'West'].Units
central = df[df.Region == 'Central'].Units
east = df[df.Region == 'East'].Units
```

#### 다중 박스플롯을 그린다:

```python
plt.boxplot([west,central,east], 0, labels = ['West','Central','East'])
plt.show()
```

```python
plt.boxplot([west, east, central], notch=False, labels=['WEST', 'EAST', 'CENTRAL'])
plt.show()
```

```python
df2 = pd.DataFrame(data={'West': list(west)})

df2.plot.box()
plt.show()
```

- 다른 방법

```python
df.groupby('Region').Units.count()
```

```python
# list(df.groupby('Region'))
```

```python
my_groups = df.groupby('Region')
my_names = []
my_list = []

for name, dfb in my_groups:
    my_names.append(name)
    my_list.append(dfb.Units)
```

```python
plt.boxplot(my_list, notch=False, labels=my_names)
plt.show()
```
