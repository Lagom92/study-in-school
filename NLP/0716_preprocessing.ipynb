## 토크나이저

```python
import nltk
import wordcloud
```

```python
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
```

```python
text = """
I'm only one call away.
I'll be there to save the day.
Superman got nothing on me, I'm only one call away.
And when you're weak, I'll be strong, I'm gonna keep holding on.
Now don't you worry it won't be long darling.
And when you feel like hope is gone just run into my arms.
"""
```

```python
print(WordPunctTokenizer().tokenize(text))
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
print(TreebankWordTokenizer().tokenize(text))
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
# 그냥 단어들마다 나눠지도록 정규표현식을 작성
re = r'\w+'
print(RegexpTokenizer(re).tokenize(text))
```

```python
from nltk.tokenize import sent_tokenize
```

```python
print(sent_tokenize(text))
```

## 어간 추출

```python
from nltk.stem import PorterStemmer
```

```python
words = WordPunctTokenizer().tokenize(text)
```

```python
for word in words:
    print(PorterStemmer().stem(word))
```

## stopword(불용어)

문장의 의미를 형성하는 것에 있어 큰 영향을 미치지 않는 단어들

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
sw = stopwords.words('english')
```

```python
print(sw)
```

```python
# words 리스트 안에서 sw에 해당하는 단어를 제외시킨 리스트 생성
bag = []
sw += ['.', ',', '`', "'"]
for word in words:
    if word.lower() not in sw and word.lower() not in bag:
        bag.append(word) 
```

```python
bag
```

```python
from collections import Counter
```

```python
# 토큰의 빈도수
count_list = Counter(bag)
print(count_list)
```

```python
# 빈도수가 가장 높은거 10개
common_cl = count_list.most_common(10)
```

```python
common_cl
```

```python
# 인덱스 부여

common_cl_dict = {}

# 빈도수 순위대로 0 - 9 위(빈도수가 가장 높은게 0등)
# 딕셔너리 형태로, '단어': 빈도순위

for idx, common in enumerate(common_cl):
    common_cl_dict[common[0]] = idx
```

```python
print(common_cl_dict)
```

```python
# 원-핫 인코딩
import numpy as np

num = len(common_cl_dict)
np.eye(num)
```

```python
# 원-핫 인코딩 방법 2
n = len(common_cl_dict)
one_hot_list = []

for val in common_cl_dict.values():
    one_hot = [0] * n
    one_hot[val] = 1
    one_hot_list.append(one_hot)       
```
