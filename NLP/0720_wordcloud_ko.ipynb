```python
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = """
모두 잠드는 밤에 혼자 우두커니 앉아
다지나 버린 오늘을 보내지 못하고서 깨어있어
누굴 기다리나 아직 할 일이 남아있었던가
그것도 아니면 돌아가고 싶은 그리운 자리를 떠올리나
무릎을 베고 누우면 나 아주 어릴적 그랬던 것처럼 머리칼을 넘겨줘요
그 좋은 손길에 까무룩 잠이 들어도 잠시만 그대로 두어요
깨우지 말아요 아주 깊은 잠을 잘 거예요
조용하던 두 눈울 다시 나에게 내리면
나 그때처럼 말갛게 웃어보일수 있을까
나 지친것 같아 이정도면 오래버틴것 같아
그대있는 곳에 돌아갈 수 있는 지름길이 있다면 좋겠어
무릎을 베고 누우면 나 아주 어릴적 그랬던 것처럼 머리칼을 넘겨줘요
그 좋은 손길에 까무룩 잠이 들어도 잠시만 그대로 두어요
깨우지말아요 아주깊은 잠을 잘거예요
스르르르르 스르르 깊은 잠을 잘 거예요
스르르르르 스르르 깊은 잠을
"""
```

```python
words = Okt().morphs(text)
```

```python
word_cnt = Counter(words)
```

```python
words = word_cnt.most_common(50)
```

```python
stop_words = ['은', '을', '는', '어요', '에', '이', '\n']
```

```python
bag_words = {}
for word, cnt in words:
    if word not in stop_words:
        bag_words[word] = cnt
```

```python
wc = WordCloud(background_color='white', font_path="c:/Windows/Fonts/malgun.ttf")
wc.generate_from_frequencies(bag_words)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
