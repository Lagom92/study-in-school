# matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## plt.plot()

```python
x = [1, 2, 3]
y = [2, 4, 6]
plt.plot(x, y)
plt.show()
```

```python
# Line Style
# '-', '--', '-.', ':''
plt.plot(x, y, ls="--")
plt.show()
```

```python
# Marker Style
# ., o, v, ^, <, >, s, p, 1, 2, 3, 4, *, h, H, +, D, _
plt.plot(x, y, marker='o')
plt.show()
```

```python
# Line weight
plt.plot(x, y, lw='4')
plt.show()
```

```python
# Line color
plt.plot(x, y, color='r')
plt.show()
```

```python
plt.plot(x, y, marker='o', markerfacecolor='m', markersize=10)
plt.show()
```

```python
# markeredgewidth
plt.plot(x, y, marker='o', markerfacecolor='m', markersize=10, markeredgewidth=3)
plt.show()
```

```python
x = np.arange(7)
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls='--', marker='o', mec='g')
plt.show()
```

- markerfacecolor='g'

- mec='g'

## 차트 옵션

### lim

- xlim()

- ylim()

```python
# 그림 범위 조정
plt.xlim(-2, 10)
plt.ylim(-5, 15)

plt.plot(x, y, ls='--', marker='o', mec='g')
plt.show()
```

### ticks

- xticks

- yticks

```python
plt.xticks([0, 3, 6])
plt.yticks([1, 5, 9])
# 격자 무늬
plt.grid()

plt.plot(x, y, ls='--', marker='o', mec='g')
plt.show()
```

```python
# 여러개의 그래프를 그릴 수 도 있다.
a = [1, 2, 3, 4]
b = [2, 4, 6, 8]
c = [3, 6, 8, 12]

plt.plot(a)
plt.plot(b)
plt.plot(c)
plt.show()
```

```python
# 범례 사용하기
a = [1, 2, 3, 4]
b = [2, 4, 6, 8]
c = [3, 6, 8, 12]

plt.plot(a, label='a')
plt.plot(b, label='b')
plt.plot(c, label='c')

plt.legend()
plt.show()
```

## Matplotlib 연습 01

```python
drive_path = r"c:/Apps/temp/"

data = pd.read_csv(drive_path + "Traffic_Accident_2017.csv", encoding="euc-kr")
data.T
```

```python
# 요일별 사고 건수
data.요일.value_counts()
```

```python
# 위 결과를 월 ~ 일 순으로 정렬하기
# 팬시 인덱싱 사용
t = data.요일.value_counts()
y = t[['월', '화', '수', '목', '금', '토', '일']]
y
```

```python
# 선 그래프
# plot
x = range(7)

plt.plot(x, y)
plt.show()
```

```python
# 막대 그래프
# bar
x = range(7)

plt.bar(x, y)
plt.show()
```

```python
# x 축 이름 변경
# xticks()
x = range(7)

plt.xticks(x, ['월', '화', '수', '목', '금', '토', '일'])

plt.bar(x, y)
plt.show()
```

```python
# y 축 조정
# ylim
x = range(7)
plt.xticks(x, ['월', '화', '수', '목', '금', '토', '일'])

plt.ylim(450, 700)

plt.bar(x, y)
plt.show()
```

```python
# title
x = range(7)
plt.xticks(x, ['월', '화', '수', '목', '금', '토', '일'])
plt.ylim(450, 700)

plt.title("요일별 교통사고 사망 건수")

plt.bar(x, y)
plt.show()
```

```python
# label
x = range(7)
plt.xticks(x, ['월', '화', '수', '목', '금', '토', '일'])
plt.ylim(450, 700)
plt.title("요일별 교통사고 사망 건수")

plt.xlabel("요일")
plt.ylabel("사고건수")

plt.bar(x, y)
plt.show()
```

### 차 vs 차 사고다발지역(발생지 시도 시각화)

```python
data.head().T
data[['발생지시도', '사고유형_대분류']].head()
```

```python
y = data[data['사고유형_대분류'] == '차대차']['발생지시도'].value_counts()
y
```

```python
# 발생시지도 이름 가져오기
region = data.발생지시도.unique()

# y를 발생시지도 순으로 정렬하기(팬시 인덱싱 사용)
y = y[region]

x = range(len(y))
```

```python
plt.title("차 vs 차 사고다발지역")
plt.xticks(x, region)
plt.ylabel("발생 건수")
plt.xlabel("지역")

plt.bar(x, y)
plt.show()
```
